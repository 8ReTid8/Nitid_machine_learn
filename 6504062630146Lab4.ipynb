## ข้อ 1

```python
import pandas as pd
import numpy as np

df = pd.read_csv("https://gist.githubusercontent.com/omarish/5687264/raw/7e5c814ce6ef33e25d5259c1fe79463c190800d9/mpg.csv")
df
```

## ข้อ 2

```python
print(df.isin(['?']).sum())
df.dtypes
```

## ข้อ 3

```python
df = df.drop(['name'],axis=1)
df['horsepower'].replace(('?', np.nan),inplace = True)
df['horsepower'] = df['horsepower'].astype(int)
df['horsepower'].fillna(df['horsepower'].mean(), inplace = True)
print(df.isin(['?']).sum())
```

## ข้อ 4

```python
from sklearn.model_selection import train_test_split
X = df.drop(["mpg"],axis=1)
Y = df['mpg']
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
print(X_train.shape, X_test.shape)
```

## ข้อ 5

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
# print('pred :',y_pred)
print('w1 : ',model.coef_)
print('w0 : ',model.intercept_)
```

## ข้อ 6

```python
print(mean_squared_error(y_test, y_pred))
```

### อับโหลดข้อมูลจ 'transactions.csv' จาก google classroom

```python
import pandas as pd
import numpy as np
df1 = pd.read_csv("transactions.csv")
df1.dropna(axis=0,inplace = True)
# df1['Date'] = pd.to_datetime(df1['Date'])
df1['Date'] = pd.to_datetime(df1['Date'], format='%d/%m/%Y')
df1.set_index('Date',inplace = True)
df1
```

## ข้อ 8

```python
print(df1['P1'].mean())
print(df1['P3'].mean())
```

## ข้อ 9

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
dfp1 = df1.drop(['P3','P2'],axis=1)
dfp1['P1_t+1'] = df1['P1'].shift(-1)
dfp1.fillna(dfp1.mean(),inplace = True)
print(dfp1)
X = dfp1.drop(["P1_t+1"],axis=1)
Y = dfp1['P1_t+1']
X_train, X_test, y_train, y_test = train_test_split(X, Y, train_size=700,shuffle=False)
print(X_train.shape, X_test.shape)
model1 = LinearRegression()
model1.fit(X_train, y_train)
y_pred = model1.predict(X_test)
print('pred :',y_pred)
print('w1 : ',model1.coef_)
print('w0 : ',model1.intercept_)
```

## ข้อ 10

```python
print("mean absolute error (MAE): ",mean_absolute_error(y_test, y_pred))
```
